```python

import pandas as pd

# Load the dataset
df = pd.read_csv("books.csv")
cdf = df.drop_duplicates()
cdf = df.dropna()

harry_potter_books = cdf[cdf["original_title"].str.contains("Harry Potter", case=False)]

print(harry_potter_books)

harry_potter_books_cleaned = harry_potter_books.dropna()


most_selling_books = harry_potter_books_cleaned.sort_values(by="ratings_count", ascending=False)

print("Most selling books within the Harry Potter series:")
print(most_selling_books[["title", "ratings_count"]].head())

average_rating = harry_potter_books_cleaned["average_rating"].mean()

print("Average rating of the Harry Potter books:", average_rating)


```
